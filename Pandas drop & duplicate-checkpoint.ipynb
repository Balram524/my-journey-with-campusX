```python
# Assuming you have a byte sequence, let's call it 'data'
data = b'\xd5...'

try:
    # Try to decode the byte sequence using UTF-8
    decoded_data = data.decode('utf-8')
    print(decoded_data)
except UnicodeDecodeError as e:
    # Handle the decoding error
    print(f"Error decoding data: {e}")
    # You can choose to ignore the problematic byte or replace it
    # For example, replace invalid bytes with a placeholder character
    decoded_data = data.decode('utf-8', errors='replace')
    print(decoded_data)
```
