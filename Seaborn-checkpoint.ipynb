```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

```python
# plot here 
sns.relplot(x='total_bill', y='tip', data=data)
```

```python
# Plot here
sns.relplot( x='total_bill', y='tip',hue = 'smoker', data = data )
```

```python
# Plot here
sns.relplot( x='total_bill', y='tip',hue = 'smoker',style='sex', data = data )
```

```python
# Plot here
sns.relplot( x='total_bill', y='tip',hue = 'smoker',style='sex', size ='size', sizes= (15,200), data = data )
```

```python

# Plot here
sns.scatterplot( x='total_bill', y='tip',hue = 'smoker',style='sex', data = data )
```

```python
sns.catplot(x='day', y='tip', kind='strip', jitter=0, data= data)
#jitter
```

```python
sns.catplot(x='day', y ='tip',kind='swarm',data = data)
```

```python
sns.catplot(x='day', y ='tip',kind='swarm',hue = 'sex',data = data)
```

```python
sns.swarmplot(x='day', y ='tip',data = data)
```

```python
#Bar plot and count plot 
```

```python
sns.catplot(x='smoker',y = 'total_bill', kind= 'bar', data = data)
```

```python
sns.catplot(x='smoker', y='total_bill',hue= 'sex', kind = 'bar',data=data)
```

```python
#count plot  is very similar to categorial plot 
```

```python
sns.catplot(x='sex',hue= 'smoker', kind = 'count',data=data)
```

```python
#Heatmaps - A heat map is a graphical representation of data where the individual values contained in a matrix are represented as colors 
```

```python
data= sns.load_dataset('flights')
```

```python
data.head()
```

```python
sns.catplot(x='month',y='passengers',hue= 'year', kind = 'bar',data= data)
```

```python
 x=data.pivot_table(index='year', columns='month',values='passengers',aggfunc='sum')
```

```python
sns.heatmap(x, cbar='false')
```

```python
sns.heatmap(x, cbar= 'false',linewidths=0.5,annot='true',fmt='d')
```
